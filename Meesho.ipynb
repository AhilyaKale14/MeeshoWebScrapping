```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
with open("Aakarsha Petite Kurtis.html") as fp:
    content = BeautifulSoup(fp, "html.parser")
```

```python
print(content)
```

```python
data=content.find_all('div',{'class':'sc-gsDJrp Grid__Row-sc-4ki5nk-0 emKfRB fbGkYa'})
```

```python
print(data)
```

```python
listing_name=[]
listing_price=[]
listing_picture_url=[]
product_details=[]
name_of_the_seller=[]
trusted_supplier=[]
product_ratings=[]
```

```python
for items in data:
    name=items.find('span',attrs={'class':'Text__StyledText-oo0kvp-0 gpVzIL'})
    price=items.find('h4',attrs={'class':'Text__StyledText-oo0kvp-0 fka-Dwo'})
    url=items.find('img')['src']
    details=items.find('div',attrs={'class':'Card__BaseCard-b3n78k-0 WvMul ProductDescription__DetailsCardStyled-sc-1tmqkzf-0 eHfqpm ProductDescription__DetailsCardStyled-sc-1tmqkzf-0 eHfqpm'})
    seller=items.find('span',attrs={'class':'Text__StyledText-oo0kvp-0 fFihtW ShopCardstyled__ShopName-du9pku-6 gdULmH ShopCardstyled__ShopName-du9pku-6 gdULmH'})
    supplier=items.find('p',attrs={'class':'Text__StyledText-oo0kvp-0 hYxNyq'})
    ratings=items.find('h1',attrs={'class':'Text__StyledText-oo0kvp-0 bmJhiP'})
    listing_name.append(name)
    listing_price.append(price)
    listing_picture_url.append(url)
    product_details.append(details)
    name_of_the_seller.append(seller)
    trusted_supplier.append(supplier)
    product_ratings.append(ratings)
```
